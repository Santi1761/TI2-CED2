```python
import matplotlib.pyplot as plt

sizes = ["10^2", "10^3", "10^4", "10^5", "10^6"]
average = [18.6, 3.0, 26.0, 68.2, 398.8]

plt.plot(sizes, average, marker='o', linestyle='-', color='blue')
plt.xlabel('Datas size')
plt.ylabel('Average (milliseconds)')
plt.title('Average in function of Datas size')
plt.show()
```
